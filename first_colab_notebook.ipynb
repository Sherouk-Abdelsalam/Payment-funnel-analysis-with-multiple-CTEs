```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
plt.style.use('ggplot')
pd.set_option('display.max_columns', 5614615)
df= pd.read_csv('coaster_db.csv')
df.head(2)
```

```python
print('Shape:\n\t', df.shape)
print()
print('Columns:\n\t', df.columns)
print()
print('Data types:\n\t',  df.dtypes)
print()
print('Describtion')
df.describe()
```

```python
df = df.rename(columns = {'coaster_name' : 'Coaster_Name',
                   'year_introduced': 'Year_Introduced',
                    'latitude': 'Latitude',
                    'longitude':'Longitude',
                    'speed_mph' :  'Speed_mph',
                    'height_ft': 'Height_ft',
                    'opening_date_clean' : 'Opening_Date',
                    'Inversions_clean' : 'Inversions',
                    'Gforce_clean': 'Gforce'})
```

```python
df = df[['coaster_name',
    #'Length', 'Speed',
    'Location',  'Status',
    #'Opening date',   #  'Type',
    'Manufacturer',
    #'Height restriction', 'Model', 'Height',
    #   'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
    #   'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
    #   'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
    #  'Track layout', 'Fastrack available', 'Soft opening date.1',
    # 'Closing date',    'Opened',
    #'Replaced by', 'Website',
    #  'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
    # 'Single rider line available', 'Restraint Style',
    #'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
       'year_introduced', 'latitude', 'longitude', 'Type_Main',
       'opening_date_clean',
    #'speed1', 'speed2', 'speed1_value', 'speed1_unit',
       'speed_mph',
    # 'height_value', 'height_unit',
    'height_ft',  'Inversions_clean', 'Gforce_clean']].copy()
 # Step 2: Making sure the data types are correct and changing them if not as we did with the opening date being an object
print(df.dtypes)
print()
df['opening_date_clean']=pd.to_datetime(df['opening_date_clean'])
print()
print(df.dtypes)
```

```python
print('Count of Dups:\n\t', df.duplicated().sum())
#Check the dubs of a subset
df.loc[df.duplicated(subset=['Coaster_Name','Opening_Date','Location'])]
```

```python
df.query('Coaster_Name == "Little Dipper (Conneaut Lake Park)"')
```

```python
df= df.loc[~df.duplicated(subset=['Coaster_Name','Opening_Date','Location'])] \
.reset_index(drop= True)
df
```

```python
ax = df['Type_Main'].value_counts().plot(kind='bar', title= 'Volume per type', color='grey')
ax.set_xlabel('Type')
ax.set_ylabel('Count')
plt.show()


axx = df['Year_Introduced'].value_counts() \
    .head(10) \
    .plot(kind='bar', title = 'Top 10 Years Coasters Introduced',color='grey')
axx.set_ylabel('Count')
plt.show()



ax= df['Speed_mph'].plot(kind='hist',color='grey', bins=20, title= 'Distributions of speed (mph)')
ax.set_xlabel('Speed (mph)')
ax.set_ylabel('Frequency')
plt.show()
```

```python
sns.scatterplot(x='Speed_mph',
                y='Height_ft',
                hue='Type_Main',
                data=df)
plt.show()


sns.pairplot(df, vars= ['Year_Introduced', 'Speed_mph','Height_ft','Gforce'])
plt.show()
```

```python
ax= df.query('Location!= "Other"')\
    .groupby('Location')['Speed_mph']\
    .agg(['mean','count'])\
    .query('count>=10')\
    .sort_values('mean')['mean']\
    .plot(kind='barh',title='Average coaster speed by location',color= 'gray')
```
